```python
def create_df(responses):
    
    real_state_id = []
    item_id = []
    code = []
    neighborhood = []
    price = []
    surface = []
    description = []
#     all_data_frames = []
    
    soup = BeautifulSoup(responses[0].content, 'html.parser')
    scripts = soup.extract('script')
    script_tag = soup.find('script', text=re.compile(r'window\.__INITIAL_PROPS__'))
    text = script_tag.text

    text = text.replace('\\', '')
    text = text.replace('u00ED', 'i')
    text = text.replace('u00F3', 'o')
    text = text.replace('u00F1', 'n')
    text = text.replace('u00FAnico', 'u')
    text = text.replace('u00BF', '')

    pattern_id = r'"realEstateAdId":(?:null),|"realEstateAdId":"([\wd-]+)",'
    real_state_id_i = re.findall(pattern_id, text)
    real_state_id_i = real_state_id_i[1:]
    real_state_id.extend(real_state_id_i)
        
    pattern_item_id = r'"id":([\d]+),"isDiscarded"'
    item_id_i = re.findall(pattern_item_id, text)
    item_id_i = item_id_i[1:]
    item_id.extend(item_id_i)
            
    pattern_pd = r',"zipCode":(?:null),|,"zipCode":"(\d{5})",'
    code_i = re.findall(pattern_pd, text)
    code_i = code_i[1:]
    code.extend(code_i)      
        
    pattern_neighborhood = r',"upperLevel":"([\w -]+)"'
    neighborhood_i = re.findall(pattern_neighborhood, text)
    neighborhood_i = neighborhood_i[1:]
    neighborhood.extend(neighborhood_i)        
        
    pattern_price = r'"rawPrice":(?:null),|"rawPrice":([\d]+),'
    price_i = re.findall(pattern_price, text)
    price_i = price_i[1:]
    price.extend(price_i)
            
    pattern_surface = r'"surface","value":([\d]+),' 
    surface_1 = re.findall(pattern_surface, text)
    surface_1 = surface_1[1:]
    surface.extend(surface_1)     
        
    pattern_description = r',"description":(?:null),|,"description":"([\w -d.]+)'
    description_i = re.findall(pattern_description, text)
    description_i = description_i[1:]
    description.extend(description_i)

    for i in range(1, 8):
        
        soup = BeautifulSoup(responses[i].content, 'html.parser')
        scripts = soup.extract('script')
        script_tag = soup.find('script', text=re.compile(r'window\.__INITIAL_PROPS__'))
        text = script_tag.text

        text = text.replace('\\', '')
        text = text.replace('u00ED', 'i')
        text = text.replace('u00F3', 'o')
        text = text.replace('u00F1', 'n')
        text = text.replace('u00FAnico', 'u')
        text = text.replace('u00BF', '')

        pattern_id = r'"realEstateAdId":(?:null),|"realEstateAdId":"([\wd-]+)",'
        real_state_id_i = re.findall(pattern_id, text)
        real_state_id.extend(real_state_id_i)
        
        pattern_item_id = r'"id":([\d]+),"isDiscarded"'
        item_id_i = re.findall(pattern_item_id, text)
        item_id.extend(item_id_i)
            
        pattern_pd = r',"zipCode":(?:null),|,"zipCode":"(\d{5})",'
        code_i = re.findall(pattern_pd, text)
        code.extend(code_i)     
        
        pattern_neighborhood = r',"upperLevel":"([\w -]+)"'
        neighborhood_i = re.findall(pattern_neighborhood, text)
        neighborhood.extend(neighborhood_i)           
         
        pattern_price = r'"rawPrice":(?:null),|"rawPrice":([\d]+),'
        price_i = re.findall(pattern_price, text)  
        price.extend(price_i)
            
        pattern_surface = r'"surface","value":(?:null),|"surface","value":([\d]+),"maxValue":0,"minValue":0}],|"key":"rooms","value":([\d]+),"maxValue":0,"minValue":0}],"hasListLogo"' 
        surface_i = re.findall(pattern_surface, text)
        surface_i_2 = []
        for k in range(0, len(surface_i)):
            surface_i_2.append(surface_i[k][0])
        surface.extend(surface_i_2)       
        
        pattern_description = r',"description":(?:null),|,"description":"([\w -d.]+)'
        description_i = re.findall(pattern_description, text)
        description.extend(description_i)
        
        
#         try:
#             data = pd.DataFrame({'item_id': item_id, 'postal_code': code, 'neighborhood': neighborhood, 'price': price, 'surface': surface, 'description': description})
        
#         except Exception as e:
#             print(f"Error occurred while creating DataFrame for response {i}: {str(e)}")
        
#         else:
#             all_data_frames.append(data)

    data = pd.DataFrame({'item_id': item_id, 'postal_code': code, 'neighborhood': neighborhood, 'price': price, 'surface': surface, 'description': description})

    return data
#     return all_data_frames 
```
